#### Descripción

Los datos recabados de este servicio web son de mediciones cada hora de una serie de estaciones de contaminación atmosférica.

https://valencia.opendatasoft.com/explore/dataset/estacions-contaminacio-atmosferiques-estaciones-contaminacion-atmosfericas/information/

Cada estación realiza una serie de mediciones acerca de la calidad del aire, y de ciertas sustancias que se consideran importantes para analizar la contaminación atmosférica en la ciudad.

- so2: niveles de dióxido de azufre
- no2: niveles de dióxido de nitrogeno
- o3: niveles de ozono
- co: niveles de monóxido de carbono
- pm10: niveles de partículas en suspensión hasta 10um
- pm25: niveles de partículas en suspensión hasta 2,5um
- fecha_carga: timestamp de la actualización de los datos. (Sugerencia: cambiar a tipo `datetime`)
- calidad_ambiental: estimación de la calidad del aire
- fiwareid: 
- geo_shape: coordenadas geográficas
- geo_point_2d: latitud-longitud

```python
import os
import pandas as pd

ruta_csv = os.path.join("valencia_pollution_dataset.csv")
```

## 01
Carga de datos de csv (valencia_pollution_dataset.csv) en un DataFrame

```python
#Leemos el archivo usando la funcion de pandas que se encarga de ello
valencia_pollution = pd.read_csv(ruta_csv, sep=',')

display(valencia_pollution)
```

## 02
- ¿Cuál es el rango temporal del dataset? Indique las fechas de inicio y fin de esta muestra de datos.
- Obtención del número de estaciones que tienen un sensor para medir monóxido de carbono

```python
#Echamos un primer vistazo a los tipos de datos del DataFrame
valencia_pollution.info()
```

```python
#Cambiamos tipo de dato de la columna "fecha_carga" a tipo de dato datetime para poder tarbajar con el
valencia_pollution["fecha_carga"] = pd.to_datetime(valencia_pollution["fecha_carga"])
valencia_pollution.info()
```

```python
display(valencia_pollution)
```

```python
#Ahora podemos trabajar con este dato tipo datetime
fecha_inicio = valencia_pollution["fecha_carga"].min()
fecha_fin = valencia_pollution["fecha_carga"].max()
print(f"La fecha de inicio de las mediciones es {fecha_inicio}")
print(f"La fecha de fin de las mediciones es {fecha_fin}")

print(f"El rango temporal es {fecha_fin - fecha_inicio}")
```

Es decir vemos que la primera fecha es el 24 de mayo a las 20:20 y la ultima el 26 de mayo a las 10:20, siendo el rango temporal de 1 dia y 14 horas

```python
#Para ver el numero de estaciones con detectores de monoxido de carbono, aplicamos una mask y usamos otro DataFrame valencia_pollution2 para 
# no modificar el original

mask = pd.notnull(valencia_pollution["co"])
valencia_pollution2 = valencia_pollution[mask]
valencia_pollution2 = valencia_pollution2.drop_duplicates(subset="objectid")

display(valencia_pollution2)
```

```python
print(f"El numero de estaciones con medidores de monoxido de carbono es {valencia_pollution2["objectid"].count()}")
```

## 03

- Indique el número de estación con la media más alta de mediciones de monóxido de carbono

```python
#Hacemos aqui una agregacion por la columna objectid y sacamos la media. Volvemos a aplicar el mismo mask de antes que solo
#dejaba las filas que tenian algun valor de la medicion de monoxido de carbono no nulo, pero esta vez sin eliminar duplicados puesto que
#lo que queremos hacer es agregar por estacion

valencia_pollution2 = valencia_pollution[mask]
valencia_pollution_agregation = valencia_pollution2.groupby('objectid')

display(valencia_pollution_agregation.mean(numeric_only=True)['co'])

type(valencia_pollution_agregation.mean(numeric_only=True)['co'])
```

```python
indice = valencia_pollution_agregation.mean(numeric_only=True)['co'].idxmax()

print(f"La estacion con la media mas alta de mediciones de monoxido de carbono es {int(indice)}")
```

## 04
Realizar el histograma de dióxido de nitrógeno de la estación del punto anterior

```python
#Sacamos un DatFrame con las mediciones de la estacion que nos interesa
import matplotlib.pyplot as plt

mask2 = valencia_pollution['objectid'] == indice

valencia_pollution_station = valencia_pollution[mask2]
display(valencia_pollution_station)
```

```python
#Usamos la propiedad de representacion de los DataFrames con la columna del DataFrame que nos interesa
valencia_pollution_station["no2"].hist(edgecolor='black', bins = 10)
plt.xlabel("Nivel de No2")
plt.grid(False)
plt.title(f"Histograma de Dioxido de Carbono de estacion {indice}")
plt.show()
```

## 05
Realizar gráfica con  la línea temporal de las mediciones de ozono de la estación del punto anterior.

```python
#Para hacer la representacion formateamos las fechas quitando la zona horaria +00:00, que no nos interesa, usando la libreria mdates preparada para esto
#Usamos tambien la funcion plt.xticks para rotar los labels en el eje x y que no se solapen
import matplotlib.dates as mdates
fig, ax = plt.subplots()
ax.plot(valencia_pollution_station["fecha_carga"], valencia_pollution_station["o3"], 'go-', alpha=0.6)
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))

plt.xticks(rotation=45)
ax.set_xlabel("Fecha")
ax.set_ylabel("Nivel de ozono")
ax.set_title("Variacion temporal de las mediciones de ozono")
plt.show()
```
